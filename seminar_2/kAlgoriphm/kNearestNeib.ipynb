## Алгоритм k ближайших соседей

Алгоритм:
- Вычислить расстояние до каждой из точек выборки
- Выбрать k ближайших соседей
- Оценить принадлежность к классу, используя формулу: 
$$
\hat{y}_{j} = \sum_{i = 1}^{k} = w_{i}*\operatorname{I}(x_{i} \in Y_{j})
$$
> Принадлженость точки к классу $Y_{j}$

Где $w_{i}$ - вес i-ой точки(вклад в оценку принадлежности)

##### Подключение библиотек

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

##### Загрузка dataset'а

```python
irisData = pd.read_csv(r"IRIS.csv")
irisData.head()
```

```python
irisData.species.unique() # Виды ириса
```

Нумерация типов:
- <<Iris-setosa>> - 0
- <<Iris-versicolor>> - 1
- <<Iris-virginica>> - 2

```python
irisData["IrisType"] = [(t == "Iris-versicolor") + 2*(t == "Iris-virginica") for t in irisData.species]
```

Зависимость типа от разных параметров:

```python
parametrNames = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
fig, ax = plt.subplots(len(parametrNames),
                       len(parametrNames),
                       figsize = (32,18),
                       dpi = 500)

for x,a in enumerate(ax[0]):
    a.set_title(parametrNames[x], fontsize = 19)

for y,a in enumerate(ax[:,0]):
    a.set_ylabel(parametrNames[y],fontsize = 19)

for x,nameX in enumerate(parametrNames):
    for y,nameY in enumerate(parametrNames):
        ax[x,y].scatter(irisData[nameX],irisData[nameY],c = irisData.IrisType)

```

Использую данные petal_length, sepal_length для разделения красных и желтых типов(чтобы попробовать использовать KNN)
