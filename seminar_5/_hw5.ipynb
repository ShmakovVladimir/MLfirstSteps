**Описание:**

**Нормальный интервал**

Как мы знаем, $\overline{X}$ является ассимптотически нормальной оценкой математического ожидания с ассимтпотической дисперсией $\operatorname{D} X_{1}$. 

К, сожалению, дисперсия распределения исходной выборки нам не известна. Поэтому будем использовать бутстрепную оценку дисперсии $\hat{v}_{boot}$. 

Алгоритм нахождения $\hat{v}_{boot}$:
- генерируем $B$ выборок на основе исходной. Просто заполняем матрицу размера $B \times n$ произвольно выбранными элементами исходной выборки
- По каждой сгенерированной выборке считаем $\overline{X}$. Получаем столбец вида $(\overline{X}_{1}^{*}, \overline{X}_{2}^{*} \dots \overline{X}_{B}^{*})^{T}$
- Теперь можем посчитать бутстрепную оценку дисперсии для рассматриваемого парметра: $\hat{v} = <\overline{X}^{2}> - (<\overline{X}>)^{2}$
> тут $<X^{2}>$ обозначает среднее квадратов бутстрепных оценок  

**3.** Реализуйте функцию вычисления выборки оценок $\theta^*_1, \ldots \theta^*_B$ для оценки $\widehat \theta$ методом бутстрепа. Функция должна принимать на вход выборку и количество бутстрепных выборок $B$. 

```python
def estimator(sample):
    '''
    Делает необходимые преобразования с выборкой (выполняет подсчет статистики).
    Используется в функции bootstrap для вычисления оценок.
    
    :param sample: массив с числами
    :return: оценка, посчитанная по выборке
    '''
    
    return np.nanmean(sample, axis = 1)


def bootstrap(sample: np.ndarray, 
              B: int = int(1e3))->np.ndarray:
    '''
    Реализует процедуру бутстрепа.
    
    :param sample: исходная выборка, по которой будут вычисляться бутстрепные оценки
    :param B: число бутстрепных оценок
    :return: возвращает массив длины B c бутстрепными оценками
    '''

    n = len(sample)  # длина выборки
    step = 100
    resamples = [np.random.choice(sample, size = (step, n)) for _ in np.arange(0, B, step)]
    resample = np.concatenate(resamples, axis = 0)
    return estimator(resample)
```

**4.** Реализуйте три типа бутстрепных доверительных интервалы в виде функций, принимающих на вход выборку оценок, полученных с помощью бутстрепа. Реализовывать вычисление бутстрепных интервалов для подвыборок размера от 1 до n не нужно, только для всей выборки. 

```python
def boostrap_normal_confidence_interval(bootstrap_estimations: np.ndarray,
                                        sample: np.ndarray,
                                        estimator_function: tp.Callable = np.mean,
                                        alpha: float = 0.95):
    estimation_value = estimator_function(sample)
    bootstrap_variance = np.mean(np.power(bootstrap_estimations, 2)) - np.power(np.mean(bootstrap_estimations), 2)
    z = sps.norm.ppf((1 - alpha) / 2)          
    return estimation_value - z * np.sqrt(bootstrap_variance), estimation_value + z * np.sqrt(bootstrap_variance)
    

    
def boostrap_central_confidence_interval(bootstrap_estimations: np.ndarray,
                                         sample: np.ndarray,
                                         estimator_function: tp.Callable = np.mean, 
                                         alpha: float = 0.95) -> tuple:
    doubled_estimation = 2 * estimator_function(sample)
    qLeft = np.quantile(bootstrap_estimations, q = (1 + alpha) / 2)
    qRight = np.quantile(bootstrap_estimations, q = (1 - alpha) / 2)
    return doubled_estimation - qLeft, doubled_estimation - qRight
    
def boostrap_quantile_confidence_interval(bootstrap_estimations, alpha=0.95):
    qLeft = np.quantile(bootstrap_estimations, q = (1 + alpha) / 2)
    qRight = np.quantile(bootstrap_estimations, q = (1 - alpha) / 2)
    return (qLeft, qRight)
```

**5.** Для каждого признака постройте бутстрепные доверительные интервалы для $\theta = \mathsf{E} X_1$ и сравните их.

Генерация бутстрепных оценок

```python
bootstrap_estimations = {key: bootstrap(data[key]) for key in targetColumns}
```

Различные доверительные интервалы

```python
quantConfInterval = {key: boostrap_quantile_confidence_interval(bootstrap_estimations[key]) for key in targetColumns}
centralConfInterval = {key: boostrap_central_confidence_interval(bootstrap_estimations[key], data[key]) for key in targetColumns}
normalConfInterval = {key: boostrap_normal_confidence_interval(bootstrap_estimations[key], data[key]) for key in targetColumns}

intervals = {'label': ['Нормальный доверительный интервал', 'Квантильный доверительный интервал', 'Центральный доверительный интервал'],
             'color': ['pink', 'darkcyan', 'orange'],
             'data': [normalConfInterval, quantConfInterval, centralConfInterval],
             'pos': 0.1 + np.arange(3) / 3}
```

```python
fig, ax = plt.subplots(1, 3, figsize = (21, 6), dpi = 200)

for ind, key in enumerate(targetColumns):
    ax[ind].set_title(key)
    for j in range(3):
        key_interval = intervals['data'][j][key]
        yAx = np.linspace(0, 1, 100)
        if key_interval is not None:
            ax[ind].hlines(xmin = key_interval[0],
                           xmax = key_interval[1],
                           y = intervals['pos'][j],
                           color = intervals['color'][j],
                           label = intervals['label'][j])
            ax[ind].fill_betweenx(y = yAx, 
                                  x1 = key_interval[0], 
                                  x2 = key_interval[1],
                                  alpha = 0.05,
                                  color = intervals['color'][j])            
    ax[ind].set_ylim(0, 1)        
    ax[ind].legend()

fig.suptitle("Сравнение доверительных интервалов");
```
